```python
from BiasedErasure.main_code.simulate import create_file_name, Bloch_sphere_point_simulate
import numpy as np
import qec

Meta_params_keys = ['architecture', 'code', 'logical_basis', 'bias_preserving_gates', 
                'noise', 'is_erasure_biased', 'LD_freq', 'LD_method', 'SSR', 'cycles', 'ordering', 'decoder', 'circuit_type', 'printing', 'num_logicals', 'Steane_type', 'loss_decoder']
Meta_params = {key: None for key in Meta_params_keys}
    
bloch_point_params_keys =['erasure_ratio', 'bias_ratio']
bloch_point_params = {key: None for key in bloch_point_params_keys}





inputs = "--architecture=CBQC --code=Rotated_Surface --logical_basis=Z  --bias_preserving_gates=False \
--noise=biased_erasure_noise --is_erasure_biased=False \
--erasure_ratio=1.0 --bias_ratio=0.5 --phys_err_min=4e-2 --phys_err_max=40e-2 --num_p_phys=1 \
--output_dir=/Users/gefenbaranes/Documents/results --num_shots=1000 --distances=3 --LD_freq=1 \
--LD_method=SWAP --SSR=True --cycles=d --ordering=fowler --decoder=MLE --circuit_type=memory --num_logicals=1 --printing=False"


inputs = "--architecture=CBQC --code=Rotated_Surface --logical_basis=Z  --bias_preserving_gates=False \
--noise=atom_array --is_erasure_biased=False \
--erasure_ratio=0.0 --bias_ratio=0.5 --phys_err_min=1.5e-2 --phys_err_max=2.2e-2 --num_p_phys=1 \
--output_dir=/Users/gefenbaranes/Documents/results --num_shots=25000 --distances=3 --LD_freq=1 --loss_decoder=independent \
--LD_method=FREE --SSR=True --cycles=10 --ordering=fowler --decoder=MLE --circuit_type=memory --num_logicals=1 --printing=False"




inputs_list = inputs.split()
print(inputs_list)
for k, v in ((k.lstrip('-'), v) for k,v in (a.split('=') for a in inputs_list)):
    if k in Meta_params:
        Meta_params[k] = v
    elif k in bloch_point_params:
        bloch_point_params[k] = v
    elif k =='output_dir':
        output_dir = v
    elif k == 'phys_err_min':
        phys_err_min = float(v)
    elif k == 'phys_err_max':
        phys_err_max = float(v)
    elif k == 'distances':
        distances= [int(x) for x in v.split(',')]
    elif k == 'num_shots':
        num_shots = int(v)
    elif k == 'num_p_phys':
        num_p_phys = int(v)


print(f"distances = {distances}")
print(f"Meta params dictionary = {Meta_params}")
print(f"bloch point params dictionary = {bloch_point_params}")
phys_err_vec = np.round(np.linspace(phys_err_min, phys_err_max, num_p_phys),8)
filename = create_file_name(Meta_params, bloch_point_params)
print(filename)
folder=output_dir


Bloch_sphere_point_simulate(Meta_params=Meta_params, bloch_point_params=bloch_point_params, output_dir=output_dir, phys_err_vec=phys_err_vec, distances=distances, num_shots=num_shots)
```
