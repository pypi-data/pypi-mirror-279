```python
import voxel_world
import random
import numpy as np
from noise import pnoise3
from voxel_world import VoxelWorld
from IPython.display import display, Image as IPImage # Jupyter notebook
```

```python
world = np.array([[[1 if pnoise3(x / 10.0, y / 10.0, z / 10.0) > random.uniform(-0.2, 0.2) else 0 for z in range(16)] for y in range(16)] for x in range(16)], dtype=np.uint8)
resolution = 10
zoom = 2.0
viewing_angle=(random.randint(0, 90), random.randint(0, 90))

data=VoxelWorld.Animations.create_voxel_img(
    world,
    random.choice(list(VoxelWorld.themes.keys())),
    resolution=resolution,
    viewing_angle=viewing_angle,
    zoom=2.0,
    show_light_source=False,
    dark_bg=False
).getvalue()

display(IPImage(data))
```

```python
data2=VoxelWorld.Animations.create_voxel_img(
    world,
    random.choice(list(VoxelWorld.themes.keys())),
    resolution=resolution,
    viewing_angle=viewing_angle,
    zoom=2.0,
    show_light_source=False,
    dark_bg=False,
    singleton = (10, 10, 10)
).getvalue()

display(IPImage(data2))
```
